```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
history0=np.load('his0_resnet.npy',allow_pickle='TRUE').item()
history1=np.load('his1_resnet.npy',allow_pickle='TRUE').item()
history2=np.load('his2_resnet.npy',allow_pickle='TRUE').item()
history3=np.load('his3_resnet.npy',allow_pickle='TRUE').item()
```

```python
# Retrieve a list of accuracy results on training and test data
# sets for each training epoch
acc0 = np.array(history0['val_accuracy'])
acc1 = np.array(history1['val_acc'])
acc2 = np.array(history2['val_acc'])
acc3 = np.array(history3['val_acc'])

# Retrieve a list of list results on training and test data
# sets for each training epoch
loss0 = history0['val_loss']
loss1 = history1['val_loss']
loss2 = history2['val_loss']
loss3 = history3['val_loss']

# Get number of epochs
epochs = range(len(acc0))

plt.figure(figsize=(10,10), dpi=70)


# Plot training and validation accuracy per epoch
plt.plot(epochs, acc0, "-bo", label = "training", )
plt.plot(epochs, acc1, "-go", label = "fine_tuning1")
plt.plot(epochs, acc2, "-ro", label = "fine_tuning2")
plt.plot(epochs, acc3, "-yo", label = "fine_tuning3")
plt.legend(loc="lower right")
plt.title('Accuracy of resnet50 models')
plt.xlabel('epochs')

plt.savefig("resnet_accuracy.png")

plt.figure(figsize=(10,10), dpi=70)

# Plot training and validation loss per epoch
plt.plot(epochs, loss0, "-bo", label = "training")
plt.plot(epochs, loss1, "-go", label = "fine_tuning1")
plt.plot(epochs, loss2, "-ro", label = "fine_tuning2")
plt.plot(epochs, loss3, "-yo", label = "fine_tuning3")
plt.legend(loc="upper right")
plt.xlabel('epochs')
plt.title('Loss of resnet50 models')

plt.savefig("resnet_loss.png")
```
